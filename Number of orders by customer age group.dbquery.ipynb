```sql
WITH customer_ages AS (
  SELECT 
    f.order_number,
    YEAR(CURRENT_DATE()) - YEAR(c.birthdate) AS age
  FROM workspace.salesdb.fact_sales f
  JOIN workspace.salesdb.dim_customers c ON f.customer_key = c.customer_key
  WHERE c.birthdate IS NOT NULL
)
SELECT 
  CASE 
    WHEN age BETWEEN 18 AND 25 THEN '18-25'
    WHEN age BETWEEN 26 AND 35 THEN '26-35'
    WHEN age BETWEEN 36 AND 45 THEN '36-45'
    WHEN age BETWEEN 46 AND 55 THEN '46-55'
    WHEN age > 55 THEN '56+'
    ELSE 'Under 18'
  END AS age_group,
  COUNT(DISTINCT order_number) AS number_of_orders
FROM customer_ages
GROUP BY age_group
ORDER BY 
  CASE age_group
    WHEN 'Under 18' THEN 1
    WHEN '18-25' THEN 2
    WHEN '26-35' THEN 3
    WHEN '36-45' THEN 4
    WHEN '46-55' THEN 5
    WHEN '56+' THEN 6
  END
```
